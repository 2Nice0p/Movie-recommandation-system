```python
import ast                     #used to convert the string of list ot list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):         #taking only name form the genres
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
 def convert3(obj):                #taking the top three actors name of the cast column
    L = []
    counter = 0
    for i in ast.literal_eval(obj):         #taking only name form the cast
        if(counter != 3):
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)               ##taking only three values form the cast
```

```python
movies.head()
```

```python
def fetch_director(obj):                #taking the top three actors name of the cast column
    L = []
    for i in ast.literal_eval(obj):         #taking only name form the cast
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']
```

```python
movies['overview'] = movies['overview'].astype(str)
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace (" ","") for i in x])         #removing the spaces to between each ownself for not confusing our model eg - James Cameron to JamesCameron
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace (" ","") for i in x]) 
movies['cast'] = movies['cast'].apply(lambda x:[i.replace (" ","") for i in x]) 
movies['crew'] = movies['crew'].apply(lambda x:[i.replace (" ","") for i in x]) 
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()                   #by this we are converting the similiar words to the one single word
```

```python
def stem(text):                    #eg. [love,loving,loved] -> [love,love,love]
    y = []

    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
    
```
